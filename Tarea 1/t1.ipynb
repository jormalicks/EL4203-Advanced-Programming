```python
from t1 import CaminosPCB
import matplotlib.pyplot as plt

# Crear instancia de la clase
grid = CaminosPCB(10,10)

# Llamar a los métodos y medir su tiempo de ejecución
#print(grid.solve())

# Lista de métodos y sus tiempos de ejecución
solvers = ["Recursivo", "Iterativo", "Combinatorial"]
tiempos = []

# Ejecutar cada solución y almacenar los tiempos
for solver in solvers:
    grid.solve(solver=solver)
    tiempos.append(grid.solve.tiempo_ejecucion)

# Generar el gráfico
plt.figure(figsize=(8, 6))
plt.bar(solvers, tiempos, color='skyblue')
plt.title('Tiempos de ejecución para cada solución')
plt.xlabel('Método')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.savefig('tiempos_ejecucion.svg', format='svg')  # Guardar como SVG
plt.show()
```
